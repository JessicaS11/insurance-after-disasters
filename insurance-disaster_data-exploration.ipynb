# Insurance After Flood Disasters

Preliminary exploration of disaster and Google search data to explore the relationships between natural disasters and interest in various types of property insurance. Here we specifically look at searches for different types of insurance after flood disasters.

```python
#import needed libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
#import Google Trend temporal datasets and merge into a dataframe by date
#note that the use of datetime adds a day (01) to each year-month from the original data
ggl_searches = pd.DataFrame(columns=['Month'])

for file in glob.glob("./data/multiTimeline*.csv"):
    series = pd.read_csv(file, skiprows=2, header=0, parse_dates=[0], infer_datetime_format=True)
    series.rename(columns={series.columns[1]:str(series.columns[1])[:-17]}, inplace=True) #remove "US" part of column name
    ggl_searches = pd.merge(series, ggl_searches, how='outer', on='Month')
        
#rename 'Month' column
ggl_searches.rename(columns={'Month':'date'}, inplace=True)


#import FEMA disaster dataset
disasters = pd.read_csv('./data/DisasterDeclarationsSummaries.csv', parse_dates=[6,11,12], infer_datetime_format=True)
#make datetime columns naive
disasters.declarationDate = pd.to_datetime(disasters.declarationDate).dt.tz_localize(None)
disasters.incidentBeginDate = pd.to_datetime(disasters.incidentBeginDate).dt.tz_localize(None)
disasters.incidentEndDate = pd.to_datetime(disasters.incidentEndDate).dt.tz_localize(None)

#add whether or not there's been a flood disaster to the trend data
#still need to generalize to entire month instead of just first day of each month
ggl_searches['flood'] = ""

for datarow in ggl_searches.itertuples(index=True, name='Pandas'):
    if disasters.loc[np.logical_and(datarow.date>=disasters.incidentBeginDate, datarow.date<=disasters.incidentEndDate)\
                                    & (disasters.incidentType=='Flood')].any().any():
        ggl_searches.loc[datarow.Index,'flood'] = True
    else:
        ggl_searches.loc[datarow.Index,'flood'] = False
        
datarow=None
```

```python
#first plot: show data through time
#import matplotlib.dates as mdates
#import datetime

nrows=1; ncols=1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16,6), squeeze=False)

#plot and label google trend data
plot_cols = ('property insurance','renters insurance','homeowners insurance',\
            'flood insurance','insurance')
col_labels = ('Property', 'Renters', 'Homeowners', 'Flood', 'Overall')
color = ('brown','orange','gray', 'blue','black')
offset = [20,0,40,0,60]

idx = 0    
for col in plot_cols:    
    axes[0,0].plot(ggl_searches.date, ggl_searches[col]+offset[idx], marker='o', linestyle='-', color=color[idx], \
                   label=col_labels[idx]+' ('+str(offset[idx])+')')
    idx = idx + 1

#plot and label declared flood disasters
axes[0,0].bar(ggl_searches.loc[ggl_searches.flood==True,'date'], height=200, width=10, label='Flood Disasters')

#axis limits
#axes[0,0].set_xlim(2004,2020)
axes[0,0].set_ylim(0,165)

#legend
handles, labels = axes[0,0].get_legend_handles_labels()  # get handles
#    handles = [h[0] for h in handles]   # remove the errorbars
axes[0,0].legend(handles, labels, bbox_to_anchor=(1.15,1), loc=1, handletextpad=0.1)


#title and axis labels
fig.suptitle('Flood Disasters and Google Searches for Insurance', fontsize=18)
axes[0,0].set_xlabel('Year', fontsize=14)
axes[0,0].set_ylabel('Relative Interest Over Time', fontsize=14)
 
#explanatory text
fig.text(0.915,0.53, '(x) indicates vertical\noffset of data added\nfor clarity', ha='left', fontsize=12)
    
plt.show()
```

```python
#compute data for second plot
years = pd.to_datetime(np.arange(2004,2019,1),format='%Y')

#count number of flood disasters per year and compute min, max, and med search trends
num_floods=np.zeros(len(years))
trend_min = np.zeros(len(years))
trend_max = np.zeros(len(years))
trend_med = np.zeros(len(years))
for idx in range(0,len(years)):
    num_floods[idx] = ggl_searches.loc[np.logical_and(ggl_searches.date.dt.year==years[idx].year,\
                                                                   ggl_searches.flood==True),'date'].count()
    trend_vals = ggl_searches.loc[ggl_searches.date.dt.year==years[idx].year,'flood insurance'].values
    trend_min[idx] = np.nanmin(trend_vals)
    trend_max[idx] = np.nanmax(trend_vals)
    trend_med[idx] = np.median(trend_vals)
```

```python
#second plot: look for correlations between "flood insurance" searches and the number of floods that year

nrows=1; ncols=1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16,6), squeeze=False)

#plot_trends = (trend_min, trend_max, trend_med)
#col_labels = ('Minimum','Maximum','Median')
#color = ('orange','red','black')

#idx = 0    
#for trend in plot_trends:    
#    axes[0,0].scatter(num_floods, trend, marker='o', color=color[idx], label=col_labels[idx])
#    idx = idx + 1
axes[0,0].scatter(num_floods, trend_max, marker='o', color='red')

#legend
#handles, labels = axes[0,0].get_legend_handles_labels()  # get handles
#axes[0,0].legend(handles, labels, loc=2, handletextpad=0.1)


#title and axis labels
fig.suptitle('Annual Maxmimum Search Interest Relative to the Flood Quantity', fontsize=18)
axes[0,0].set_xlabel('Number of Months with Floods for a Given Year', fontsize=14)
axes[0,0].set_ylabel('Relative Search Interest for "Flood Insurance"', fontsize=14)

plt.show()

```
