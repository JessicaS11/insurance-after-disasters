# Insurance After Disasters

Preliminary exploration of disaster and Google search data to explore the relationships between natural disasters and interest in various types of property insurance.

```python
#import needed libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
#import Google Trend temporal datasets and merge into a dataframe by date
#note that the use of datetime adds a day (01) to each year-month from the original data

ggl_searches = pd.DataFrame(columns=['Month'])

for file in glob.glob("./data/multiTimeline*.csv"):
    series = pd.read_csv(file, skiprows=2, header=0, parse_dates=[0], infer_datetime_format=True)
    series.rename(columns={series.columns[1]:str(series.columns[1])[:-17]}, inplace=True) #remove "US" part of column name
    ggl_searches = pd.merge(series, ggl_searches, how='outer', on='Month')
        
#rename 'Month' column
ggl_searches.rename(columns={'Month':'date'}, inplace=True)

#import FEMA disaster dataset
disasters = pd.read_csv('./data/DisasterDeclarationsSummaries.csv', parse_dates=[6,11,12], infer_datetime_format=True)
#make datetime columns naive
disasters.declarationDate = pd.to_datetime(disasters.declarationDate).dt.tz_localize(None)
disasters.incidentBeginDate = pd.to_datetime(disasters.incidentBeginDate).dt.tz_localize(None)
disasters.incidentEndDate = pd.to_datetime(disasters.incidentEndDate).dt.tz_localize(None)

#add disasters to the trend datasets (right now just whether or not there's a disaster, not what type)
ggl_searches['disaster'] = ""
#ggl_searches['disaster'] = ""
#ggl_searches['disaster_type'] = ""

for datarow in ggl_searches.itertuples(index=True, name='Pandas'):
    #this just looks at whether or not there is a disaster during the date at all (note it does not generalize to the month)
    if disasters.loc[np.logical_and(datarow.date>=disasters.incidentBeginDate, datarow.date<=disasters.incidentEndDate)].any().any():
        ggl_searches.loc[datarow.Index,'disaster'] = True
        #this doesn't yet work because of cases where there are multiple disasters for a given month
        #ggl_searches.loc[datarow.Index,'disaster_type'] = \
        #disasters.loc[np.logical_and(datarow.date>=disasters.incidentBeginDate, datarow.date<=disasters.incidentEndDate),'incidentType']

    else:
        ggl_searches.loc[datarow.Index,'disaster'] = False
        
datarow=None

```

```python
#first plot: show data through time
import matplotlib.dates as mdates
import datetime

nrows=1; ncols=1
#fig=None; axes=None
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16,6), squeeze=False)

#plot and label google trend data
plot_cols = ('property insurance','renters insurance','homeowners insurance',\
             'fire insurance','flood insurance','insurance')
col_labels = ('Property', 'Renters', 'Homeowners', 'Fire', 'Flood', 'Overall')
color = ('brown','orange','gray','red','blue','black')
offset = [20,0,40,0,0,60]

#min_ys = []
#max_ys = []
idx = 0    
for col in plot_cols:    
    #axes[row,0].set_ylabel(col_labels[row])
#    min_ys.append(np.min(plot_data[col])-abs(perc))
#    max_ys.append(np.max(plot_data[col])+abs(perc))


    axes[0,0].plot(ggl_searches.date, ggl_searches[col]+offset[idx], marker='o', linestyle='-', color=color[idx], \
                   label=col_labels[idx]+' ('+str(offset[idx])+')')
    idx = idx + 1

#plot and label disaster data on secondary y-axis
#ax2y = axes[0,0].twinx()
#ax2y.bar(ggl_searches.loc[ggl_searches.disaster==True,'date'], height=200)
axes[0,0].bar(ggl_searches.loc[ggl_searches.disaster==True,'date'], height=200)

#axis limits
#axes[0,0].set_xlim(2004,2020)
axes[0,0].set_ylim(0,165)

#legend
handles, labels = axes[0,0].get_legend_handles_labels()  # get handles
#    handles = [h[0] for h in handles]   # remove the errorbars
axes[0,0].legend(handles, labels, bbox_to_anchor=(1.15,1), loc=1, handletextpad=0.1)


#title and axis labels
fig.suptitle('Natural Disasters and Google Searches for Insurance', fontsize=18)
axes[0,0].set_xlabel('Year', fontsize=14)
axes[0,0].set_ylabel('Relative Interest Over Time', fontsize=14)
    
plt.show()
```
